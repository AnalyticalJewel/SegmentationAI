```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('step_1_Preprocessing.csv')
```

```python
# Step 1: Shortlisting columns
X = data[['Total Spend', 'Age']]
```

```python
# Step 2: Scaling 
s= StandardScaler()
X_after_scaling = s.fit_transform(X)
```

```python
# Step 3: Apply K-Means Clustering (Machine Learning Algorithm)
kmeans = KMeans(n_clusters=3, random_state=42)  # You can choose the number of clusters (e.g., 3)
data['Cluster'] = kmeans.fit_predict(X_after_scaling)
```

```python
# Step 4: Visualization 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Total Spend', hue='Cluster', data=data, palette='viridis')
plt.title('Customer Segmentation Based on Total Spend and Age')
plt.xlabel('Age')
plt.ylabel('Total Spend')
plt.show()
```

```python
# The number of customers in each cluster
Customers_in_each_cluster = data['Cluster'].value_counts()
print("Number of samples in each cluster:")
print(Customers_in_each_cluster)
```
