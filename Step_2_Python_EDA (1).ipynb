```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('step_1_Preprocessing.csv')
```

```python
data.head(1)
```

```python
plt.figure(figsize=(20, 12))
sns.histplot(data['Age'], bins=range(19, 71), stat='count')  # Customize bins and use 'stat'

# Adjust x-axis tick labels to show age numbers
plt.xticks(range(19, 71))  # Set tick positions for each age

# Display the age distribution (labels might differ slightly)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.grid(True)  # Add gridlines for better readability

plt.show()
```

```python
# 3. Visualizing Purchase Amount by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Purchase Amount (USD)', data=data)
plt.title('Purchase Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
# Exclude the 'Customer ID' column
columns_to_include = data.drop(['Customer ID','Total Spend','Average Spend','Frequency of Purchases','Purchase Frequency','Customer Lifetime Value','Loyalty Score'],axis=1)

# 4. Correlation heatmap without 'Customer ID'
plt.figure(figsize=(13, 9))
sns.heatmap(columns_to_include.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
